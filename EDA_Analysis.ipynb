```python
import pandas as pd

# Load the Titanic dataset
url = 'https://raw.githubusercontent.com/datasciencedojo/datasets/master/titanic.csv'
data = pd.read_csv(url)

# Display the first few rows of the dataset
data.head()
```

```python
# Check for missing values
data.isnull().sum()
```

```python
# Impute missing values
data['Age'].fillna(data['Age'].median(), inplace=True)
data['Embarked'].fillna(data['Embarked'].mode()[0], inplace=True)

# Drop columns that are not needed
data.drop(['Cabin', 'Ticket'], axis=1, inplace=True)
```

```python
# Convert 'Sex' and 'Embarked' to categorical types
data['Sex'] = data['Sex'].astype('category')
data['Embarked'] = data['Embarked'].astype('category')
```

```python
# Descriptive statistics
data.describe(include='all')
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

# Plot distribution of Age
sns.histplot(data['Age'], kde=True)
plt.title('Distribution of Age')
plt.show()

# Plot distribution of Fare
sns.histplot(data['Fare'], kde=True)
plt.title('Distribution of Fare')
plt.show()
```

```python
# Count plot for 'Survived' feature
sns.countplot(x='Survived', data=data)
plt.title('Survival Counts')
plt.show()

# Count plot for 'Embarked' feature
sns.countplot(x='Embarked', data=data)
plt.title('Embarked Counts')
plt.show()
```

```python
# Box plot of Age by Survived
sns.boxplot(x='Survived', y='Age', data=data)
plt.title('Age vs. Survival')
plt.show()

# Violin plot of Fare by Survived
sns.violinplot(x='Survived', y='Fare', data=data)
plt.title('Fare vs. Survival')
plt.show()
```
