```python
import pandas as pd
import matplotlib.pyplot as plt

# Load your dataset
df = pd.read_csv('/content/Metadata_Country_API_SP.POP.TOTL_DS2_en_csv_v2_3401680.csv')

# Check the columns to confirm
print(df.columns)

# Calculate the distribution of countries by region
region_counts = df['Region'].value_counts()

# Plot the distribution of countries by region
plt.figure(figsize=(10, 6))
plt.bar(region_counts.index, region_counts.values, color='skyblue', edgecolor='black')
plt.xlabel('Region')
plt.ylabel('Number of Countries')
plt.title('Number of Countries by Region')
plt.xticks(rotation=90)  # Rotate x-axis labels for better readability
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt

# Load your dataset
df = pd.read_csv('/content/Metadata_Country_API_SP.POP.TOTL_DS2_en_csv_v2_3401680.csv')

# Check the columns to confirm
print(df.columns)

# Calculate the distribution of countries by income group
income_group_counts = df['IncomeGroup'].value_counts()

# Plot the distribution of countries by income group
plt.figure(figsize=(10, 6))
plt.bar(income_group_counts.index, income_group_counts.values, color='lightgreen', edgecolor='black')
plt.xlabel('Income Group')
plt.ylabel('Number of Countries')
plt.title('Number of Countries by Income Group')
plt.xticks(rotation=45)  # Rotate x-axis labels for better readability
plt.show()
```

```python
import matplotlib.pyplot as plt

# Check if the 'Gender' column exists
if 'Gender' in df.columns:
    # Calculate the gender distribution
    gender_counts = df['Country Code'].value_counts()

    # Plot the gender distribution as a bar chart
    plt.figure(figsize=(8, 6))
    plt.bar(gender_counts.index, gender_counts.values, color=['blue', 'pink'])
    plt.xlabel('Co')
    plt.ylabel('Count')
    plt.title('Gender Distribution in the Population')
    plt.show()
else:
    print("The dataset does not contain a 'Gender' column.")
```
